## loc, iloc 함수
두 함수를 사용하여 원하는 행 또는 열에 인덱싱, 슬라이싱이 가능하다.
- loc[행인덱스명] -> row조회
- loc[행인덱스명, 컬럼명] -> row&column 조회

- iloc[행인덱스번호] -> row조회
- iloc[행인덱스번호, 열인덱스번호] -> row&column조회


````python
sample.head()
````

````python
sample.loc[20170226:20190222] # 라벨 기반은 양쪽 끝값을 모두 포함한다.
````

- 보통 : 연산자는 파이썬 리스트나 넘파이 배열에서 오른쪽 끝을 포함하지 않는 게 맞음.
- 하지만 !!! Pandas의 loc[] 슬라이싱은 다르다. 양쪽 끝값을 모두 포함!!

- Pandas에서는 loc 슬라이싱이 라벨 기반이기 때문에, 우리가 라벨을 지정할 때는 직관적으로 끝도 포함되길 기대하는 경우가 많기 때문. 특히 시간 데이터에서 구간을 지정할 때 편함.

| 연산자             | 포함 여부                              |
|--------------------|-----------------------------------------|
| `iloc[start:end]`  | end 포함 X (정수 기반, 파이썬 방식)     |
| `loc[start:end]`   | end 포함 O (라벨 기반, 판다스 방식)     |


````python
sample.iloc[1:3]
````

````python
sample.loc[20180107]
````

````python
sample.loc[20170226:20180317,'남자10~14세']
````

````python
sample.iloc[:3,5]
````

## 새로운 열 추가

````python
# 새로운 '인구밀집지'라는 column추가.

# 1. 500개의 값을 리스트로 전달하기 힘들다. 일단 값을 모두 True로 지정하자.
sample['인구밀집지'] = True
sample['인구밀집지']
````

````python
# 2. 생활인구합계가 1000이상인 곳만 인구밀집지 True로 설정하자
sample['인구밀집지'] = sample['생활인구합계'] > 1000
sample['인구밀집지']
````

````python
# 3. 생활인구합계가 1000을 넘는다면 인구밀집지에 O를, 안넘는다면 인구밀집지에 X를 넣어라. => 틀림!!!
sample.apply(lambda x : x['인구밀집지'] = 'O' if x['생활인구합계'] > 1000 else 'X')
````

틀린이유
- apply(lambda x: ...) 내부에서는 값을 직접 할당하는 것이 아니라, 값을 반환(return) 해야 합니다.

- axis=1을 명시해야 행(row) 안의 요소 간에 apply 함수가 적용됩니다.

- 그 후, sample['인구밀집지']에 새로운 값을 할당하여 데이터프레임을 업데이트합니다.

````python
sample['인구밀집지'] = sample.apply(lambda x :  'O' if x['생활인구합계'] > 1000 else 'X',axis=1)
sample['인구밀집지']
````

### 마지막 행 밑에 행추가 by concat

concat()을 사용할 때는 기존 DataFrame과 추가할 DataFrame의 column(열) 이름이 같아야 합니다.

📌 컬럼이 다르면 어떻게 될까?

concat()은 자동으로 컬럼을 합집합(Union) 형태로 정렬함.

df에는 B가 있지만 new_row에는 없으면, B 값이 NaN으로 채워짐.

new_row에는 C가 있지만 기존 df에는 없으면, C 값도 일부 NaN이 됨.

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
new_row = pd.DataFrame()  # 새로운 행을 데이터프레임으로 만듦
df = pd.concat([df, new_row], ignore_index=True)

print(df)
````

# apply함수
## DataFrame에 파이썬 함수를 적용할 수 있다.


````python
# 가로 세로 높이를 이용해 부피라는 컬럼을 추가해보자
````

````python
df
````

````python
def volume(a,b,c):
    return a*b*c
````

````python
df['부피'] = df.apply(lambda x : volume(x['가로'],x['세로'],x['높이']))
#오류나는 이유: df.apply의 디폴트 axis=0.
````

x['가로'], x['세로'], x['높이']와 같이 개별 **요소!!!**에 접근하기 위해 행을 기대하고 있습니다.

=> df.apply에 axis=1으로 넣어줘야함. => **두번째 대괄호 안에 있는 것들의 요소들을 넘겨준다.**

````python
df['부피'] = df.apply(lambda x : volume(x['가로'],x['세로'],x['높이']),axis=1)
# 대부분의 경우 axis=1
df
````

````python
# 더 효율적인 방법
df['부피'] = df['가로'] * df['세로'] * df['높이']
df
````

# 중복 데이터 처리
## 중복 데이터 찾기: duplicated

````python
df = pd.DataFrame({
    'brand':['Yum','Yum','Indomie','Indomie','Indomie'],
    'style':['cup','cup','cup','pack','pack'],
    'rating':[4,4,3.5,15,5]
})
df
````

````python
df.duplicated() # 기본 keep = 'first'
````

### keep 매개변수
=> 어떤 값을 중복으로 여길지를 결정

- keep='first' => 첫번째껀 남기고. 뒷부분에 나오는걸 중복 True로.
- keep='last' => 마지막껀 남기고. 앞부분에 또 나오는걸 중복 True로.
- keep=False => 아무것도 남기지 않고, 중복된걸 모두 중복처리

````python
df.duplicated(keep=False)
````

````python
df.duplicated(keep='first')
````

````python
df.duplicated(keep='last')
````

### subset 매개변수 : 특정 컬럼만 중복을 확인할 때

````python
df
````

````python
df.duplicated(subset='brand',keep=False)
````
